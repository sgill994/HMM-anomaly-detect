```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("pca_loadfings.csv")
```

```python
fig, ax = plt.subplots()

scores = abs(df[df.columns[1:]].to_numpy().round(2))
im = plt.imshow(scores)

#ticks for variables
ax.set_yticks(np.arange(len(df["Unnamed: 0"])),labels = df["Unnamed: 0"])

#ticks for PCS
ax.set_xticks(np.arange(len(df.columns[1:])),labels= df.columns[1:])

plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")
# Loop over data dimensions and create text annotations.
for i in range(len(df["Unnamed: 0"])):
    for j in range(len(df.columns[1:])):
        text = ax.text(j, i, scores[i, j],
                       ha="center", va="center", color="w")
ax.set_title("PCA Loading Scores")
```
